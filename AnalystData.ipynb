```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as pyoff
import plotly.graph_objs as go
```

ĐỌC DỮ LIỆU TỪ FILE CSV

```python
pth = 'Du_lieu_dt_3.csv'
df = pd.read_csv(pth)
df
```

XỬ LÍ DATA TYPE CỦA SERIES 

```python
df['Gia'] = df['Gia'].str.replace('.' , '')
df['Gia'] = df['Gia'].astype('float')
df = df.round(decimals=0)
df['Gia'] = df['Gia'].astype('int64')
```

SẢN PHẨM CÓ GIÁ CAO NHẤT (1e7 = 10000000)

```python
price = pd.Series(df['Gia'])
name = pd.Series(df['Ten san pham'])

fig, ax = plt.subplots(figsize=(15, 4))

ax.plot(name , price)

ax.set_title('Bieu do the hien gia theo san pham')
ax.set_xlabel('Ten san pham')
ax.set_ylabel('Gia')
plt.xticks(name ,  rotation = 'vertical' , fontsize = 10)

plt.show()
```

```python
plot_data = [
    go.Scatter
    (
        x = df['Ten san pham'],
        y = df['Gia']
    )
]

plot_layout = go.Layout(
    xaxis = {"type":"category"},
    title = "Biểu đồ thể hiện giá theo từng sản phẩm",
    xaxis_title = 'Tên sản phẩm',
    yaxis_title = 'Giá sản phẩm'
)
fig = go.Figure(data = plot_data , layout = plot_layout)
pyoff.plot(fig)
```

NHỮNG SẢN PHẨM CÓ ĐÁNH GIÁ TỐT NHẤT

```python
plot_data = [
    go.Scatter
    (
        x = df['Ten san pham'],
        y = df['Danh gia']
    )
]

plot_layout = go.Layout(
    xaxis = {"type":"category"},
    title = "Biểu đồ thể hiện đánh giá theo từng sản phẩm",
    xaxis_title = 'Tên sản phẩm',
    yaxis_title = 'Đánh giá'
)
fig = go.Figure(data = plot_data , layout = plot_layout)
pyoff.plot(fig)
```

DOANH THU THEO TỪNG NHÃN HÀNG

```python
df['Sales']=df['Gia']*df['So Luong']
sale_values = df.groupby(by = 'Nhan hieu').agg({'Sales': 'sum'})

plot_data = [
    go.Scatter
    (
        x = sale_values.index,
        y = sale_values.get('Sales'),
    )
]

plot_layout = go.Layout(
    xaxis = {"type":"category"},
    title = "Tổng doanh thu theo từng nhãn hàng",
    xaxis_title = "Nhãn hàng",
    yaxis_title = "Sale"
)
fig = go.Figure(data = plot_data , layout = plot_layout)

pyoff.plot(fig)
```

Task: Find the best-selling and worth-buying phones in the price range of 8 million VND, in which phone sell the most quantity

```python
df2 = df
df2_tmp = df2.loc[(df2['Gia'] < 8000000) & (df2['Danh gia'] >= 4) & (df2['Gia'] > 5000000)]
most_brn = df2_tmp.loc[:,['Ten san pham','Nhan hieu','Gia','Danh gia','So Luong']]
most_brn.sort_values(by = 'So Luong' , ascending = False , inplace = True)
most_brn
```

```python
tm_mst = most_brn.groupby(by = 'Ten san pham').agg({'So Luong': 'max'}).sort_values(ascending = True , by = 'So Luong')
tm_mst.get('So Luong')
```

```python

plot_data = [
    go.Scatter
    (
        x = tm_mst.get('So Luong'),
        y = tm_mst.index
    )
]

plot_layout = go.Layout(
    xaxis = {"type":"category"},
    title = "Best selling phone in the price range of 8 million VND",
    xaxis_title = "Tên sản phẩm",
    yaxis_title = "Số lượng"
)
fig = go.Figure(data = plot_data , layout = plot_layout)

pyoff.plot(fig)
```

```python
#Phân khúc tầm trung : trên 6 triệu, dưới 10 triệu
import pandas as pd
pth = 'Du_lieu_dt_3.csv'
df = pd.read_csv(pth)


pd.unique(df["Chip"])

col_list = ['Processors','Rank']
df_rank = pd.read_csv('Smartphone-Processors-Rating.csv', usecols=col_list,index_col=0,squeeze=True)
rank_dict = df_rank.to_dict()


temp = []
for i in range(len(df)):
    phone_price = ((df["Gia"][i]).replace(".","")).replace("₫","")
    result = int(phone_price)
    temp.append(result)

df['Price'] = temp

df = df.loc[df['Price'] > 6000000]
df = df.loc[df['Price'] <= 10000000]
df = df.reset_index()

chip_column = df["Chip"]
price_column = df["Price"]


xep_hang = []

for i in range(len(df)):
    chip_rank = chip_column[i]
    phone_price = price_column[i]
    rank = rank_dict.get(chip_rank)
    if rank == None:
        rank = 0
    result = rank * int(phone_price)
    xep_hang.append(result)

df['Hieu nang / Gia'] = xep_hang

df2 = df.loc[df['Hieu nang / Gia'] > 0 ]

df2 = df2.drop(columns= ["Price"])

(df2.sort_values(['Hieu nang / Gia'])).head(5)
```
